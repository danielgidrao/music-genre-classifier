# 01 - Exploratory Analysis (FMA)

Este notebook faz EDA com `tracks.csv` e também mostra o Caminho A (features pré-computadas).

```python
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent
sys.path.append(str(PROJECT_ROOT / 'src'))

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from config import PATH_A_FEATURES_CSV
from load_data import load_tracks_metadata, load_precomputed_features, build_path_a_dataframe
```

```python
tracks = load_tracks_metadata()
tracks.shape
```

```python
genre_counts = tracks[('track', 'genre_top')].dropna().value_counts()
plt.figure(figsize=(10,4))
genre_counts.plot(kind='bar')
plt.title('Distribuição de gêneros (genre_top)')
plt.ylabel('Quantidade de faixas')
plt.tight_layout()
plt.show()
```

```python
features = load_precomputed_features()
df_path_a = build_path_a_dataframe(tracks, features, output_path=PATH_A_FEATURES_CSV)
df_path_a.head()
```

```python
df_path_a['genre_top'].value_counts().plot(kind='bar', figsize=(8,3), color='teal')
plt.title('Distribuição de gêneros no Caminho A (fma_small)')
plt.tight_layout()
plt.show()
```
